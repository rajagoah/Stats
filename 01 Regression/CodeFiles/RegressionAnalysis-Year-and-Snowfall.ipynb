```python
#importing pandas
import pandas as pd
import configparser
from configparser import ConfigParser
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt

#path of config file
path = 'C:\\Users\\aarajago\\OneDrive - Capgemini\\aakarsh personal\\6 Stats\\01 Regression\\ConfigFiles\\VariableConfig.ini'

#initialising a parser
parser = ConfigParser()
parser.read(path)

#loading data from the config parser to variables
src_file_path=parser.get('PATHS','src_file_path')

#initialise a dataframe
year_snowfall = pd.read_csv(src_file_path)

#separating the dataframe columns to 2 different variables
x = year_snowfall['Year']
y = year_snowfall['Amount (Inches)']

#sns.set_theme(color_codes=True)
%matplotlib inline
sns.regplot(x=x, y=y, data=year_snowfall);



```
